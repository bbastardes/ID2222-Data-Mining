```scala
import scala.io.Source
implicit def bool2int(b:Boolean) = if (b) 1 else 0

val source = Source.fromFile(data)
val singletons = Array.fill(100000)(0)

for (line <- source.getLines().map(x => x.split(" ").map(item => item.toInt))) {
    for (i <- 0 to 999){
        singletons(i) = singletons(i) + List(i).forall(line.toList.contains)
    }
           
}
source.close()
     
val frequent_singl = singletons.zipWithIndex.filter(x => (x._1 >= support)).map(item => item._2)
```

```scala
import scala.collection.mutable.ListBuffer
val candidates = new ListBuffer[(Int, Int)]()

for (i <- 0 to frequent_singl.size-1){
    for (j <- i+1 to frequent_singl.size-1){
        candidates += ((frequent_singl(i), frequent_singl(j)))
    }
}
```

```scala
val doubletons = Array.fill(candidates.size)(0)
val n = frequent_singl.size
val source = Source.fromFile(data)

for (line <- source.getLines().map(x => x.split(" ").map(item => item.toInt))) {
    println("Computing line...")
    for (i <- 1 to 2){
        for (j <- i+1 to frequent_singl.size-1){
            val k = ((i-1)*(n - (i/2) - 1)) + j - i - 1
            //println(i,j,k)
            val candidate = List(frequent_singl(i), frequent_singl(j))
            doubletons(k) = doubletons(k) + candidate.forall(line.toList.contains)
        }
     }       
 }
source.close()
     
//val frequent_doubl = doubletons.zipWithIndex.filter(x => (x._1 >= support)).map(item => item._2)
```
