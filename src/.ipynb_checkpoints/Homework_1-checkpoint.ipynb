```python
class Shingling():

    def __builtin__():
        x = prepreocessing(document)
        shingles = k_shingles(document, k)
        hash_val = []

        for shingle in shingles:
            hash_val.append(hash(shingle))
      
        return hash_val
    

    def prepreocessing(document):
        
        x = document.lower()
        x = x.replace(',', '')
        x = x.replace('.', '')
        x = x.replace('?', '')
        x = x.replace('!', '')
        
        return x

    def k_shingles(document, k):
        shingles = []

        for i in range(len(document)-k+1):
            shingles.append(document[i:i+k])

        return shingles


    def hash(shingle):
        h = 7

        for c in shingle:
          h = ((h << 3) + h) + ord(c)

        return h
```
